```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
train_labels = []
train_samples = []
```

```python
for i in range(50):
    # The ~5% of younger individuals who did experience side effects
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1)

    # The ~5% of older individuals who did not experience side effects
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(0)

for i in range(1000):
    # The ~95% of younger individuals who did not experience side effects
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)

    # The ~95% of older individuals who did experience side effects
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)
```

```python
train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
train_labels, train_samples = shuffle(train_labels, train_samples)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1))
```

```python
model = Sequential([
    Dense(units=16, input_shape=(1,), activation='relu'),
    Dense(units=32, activation='relu'),
    Dense(units=2, activation='softmax')
])
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x=scaled_train_samples, y=train_labels,validation_split=0.1, batch_size=10, epochs=30, verbose=2)
```

```python
#Generate new C code
import keras
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy

#load model

dense = ""

model = keras.models.load_model('crash.h5')
a = model.weights
print(model.weights)
number_of_weights = int(len(a)/2)
print("Number of Dense layers: {}".format(number_of_weights))

file_dense = open('model_dense.c','w')
file_bias  = open('model_bias.c','w')
file_header = open('model.h','w')
file_dense.write("#include \"model.h\"\n\n")
file_bias.write("#include \"model.h\"\n\n")

file_infernece = open('model_inference.c','w')


file_dense.write("Layers_t dense_layers[{}] = {}\n".format(number_of_weights,'{'))
dense_header = "extern Layers_t dense_layers[{}];\n".format(number_of_weights,'{')

file_bias.write("Layers_t bias_layers[{}] = {}\n".format(number_of_weights,'{'))
counter = 0

for num in range(number_of_weights):
    (rows,cols) = a[counter].shape
    if num == number_of_weights-1:
        file_dense.write("{} {},{} {}{};\n\n".format('{',rows,cols,'}','}'))
    else:
        file_dense.write("{} {},{} {},\n".format('{',rows,cols,'}'))        
    rows = 1
    if num == number_of_weights-1:
        file_bias.write("{} {},{} {}{};\n\n".format('{',rows,cols,'}','}'))
    else:
        file_bias.write("{} {},{} {},\n".format('{',rows,cols,'}'))
        
    counter = counter +2

counter = 0

file_infernece.write("#include <stdint.h>\n\n\n")
file_infernece.write("#include \"model.h\"\n\n\n")
file_infernece.write("static float relu(float value)\n")
file_infernece.write("{}\n".format('{'))
file_infernece.write("\tif (value > 0.0f)\n")
file_infernece.write("\t{}\n".format('{'))
file_infernece.write("\t\treturn(value);\n")
file_infernece.write("\t{}else\n".format('}'))
file_infernece.write("\t{}\n".format('{'))
file_infernece.write("\t\treturn 0;\n")
file_infernece.write("\t{}\n".format('}'))
file_infernece.write("{}\n\n".format('}'))

file_header.write("#ifndef __MODEL_H__\n")
file_header.write("#define __MODEL_H__\n")
file_header.write("#include <stdint.h>\n\n")
file_header.write("typedef struct {}\n".format('{'))
file_header.write("\tuint32_t rows;\n")
file_header.write("\tuint32_t cols;\n")
file_header.write("{} Layers_t;\n\n".format('}'))
file_header.write(dense_header)




for num in range(number_of_weights):
    (rows,cols) = a[counter].shape
    #file_dense.write("{} {},{} {},\n".format('{',rows,cols,'}'))
    file_dense.write('const float Layer_{}[{}][{}] = {}\n'.format(num,rows,cols,'{'))
    file_header.write("extern const float Layer_{}[{}][{}];\n".format(num,rows,cols))

    file_infernece.write("float * f_Layer_{}(uint32_t rows, uint32_t cols, float data[]){}\n".format(num,'{'))
    file_header.write("float * f_Layer_{}(uint32_t rows, uint32_t cols, float data[]);\n".format(num))

    file_infernece.write("\tstatic uint32_t m,n;\n")
    file_infernece.write("\tstatic float total_layer[{}];\n".format(cols))
    file_infernece.write("\tm = 0;\n")
    file_infernece.write("\tn = 0;\n")
    file_infernece.write("\tfor (n = 0; n < cols; n++)\n")
    file_infernece.write("\t{}\n".format('{'))
    file_infernece.write("\t\tfor (m = 0; m < rows; m++)\n")
    file_infernece.write("\t\t{}\n".format('{'))
    #BODY
    file_infernece.write("\t\t\ttotal_layer[n] += data[m] * Layer_{}[m][n];\n".format(num))    
    #BODY
    file_infernece.write("\t\t{}\n".format('}'))
    file_infernece.write("\t\ttotal_layer[n] += 1 * Bias_{}[1][n];\n".format(num))  
    file_infernece.write("\t\ttotal_layer[n] = relu(total_layer[n]);\n")
    file_infernece.write("\t{}\n".format('}'))
    file_infernece.write("\treturn(total_layer);\n")    
    file_infernece.write("{}\n\n".format('}'))
    

    
    for i in range(rows):
        file_dense.write('{')
        for m in range(cols):
            file_dense.write('{}'.format(a[counter][i][m]))
            if m == cols-1 and i == rows-1:
                file_dense.write('}\n};\n\n')            
            elif m == cols-1:
                file_dense.write('},\n') 
            else:
                file_dense.write(',')
    rows = 1
    print("The are {} rows and {} colmuns".format(rows,cols))
    file_bias.write('const float Bias_{}[{}][{}] = {}\n'.format(num,rows,cols,'{'))
    file_header.write("extern const float Bias_{}[{}][{}];\n".format(num,rows,cols))
    for i in range(rows):
        file_bias.write('{')
        for m in range(cols):
            file_bias.write('{}'.format(a[counter+1][m]))
            if m == cols-1 and i == rows-1:
                file_bias.write('}\n};\n\n')            
            elif m == cols-1:
                file_bias.write('},\n') 
            else:
                file_bias.write(',')

    counter = counter +2
    

    
file_dense.close()
file_bias.close()
file_header.write("#endif\n")
file_header.close()
file_infernece.close()





```
